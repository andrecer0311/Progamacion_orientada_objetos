```python
class Person:

  # Add an __init__ method
  def __init__(self, nom, edad):
    self.nom = nom
    self.edad = edad

  # Add a greet method
  def greet(self):
    print("Hola, mi nombre es " + self.nom + " y tengo " + self.edad + " años de edad")

# create an object p1
p1 = Person("Camilo", "18")

#call the greet method
p1.greet()
```
