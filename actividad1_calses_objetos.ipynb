```python
# 1. DEFINICIÓN DE LA CLASE (El Molde)
class Libro:
    def __init__(self, titulo, autor):
        # Atributos definidos por el usuario
        self.titulo = titulo
        self.autor = autor
        # Atributo con valor por defecto
        self.estado = "Disponible"

    # Método para mostrar la información actual
    def mostrar_info(self):
        print(f"--- Ficha de Biblioteca ---")
        print(f"Título: {self.titulo}")
        print(f"Autor:  {self.autor}")
        print(f"Estado: {self.estado}")


    # Método para cambiar el estado del libro (Acción)
    def prestar(self):
        if self.estado == "Disponible":
            self.estado = "Prestado"
            print(f"Has retirado el libro: '{self.titulo}'")
        else:
            print(f" Lo sentimos, '{self.titulo}' ya se encuentra prestado.")

# 2. IMPLEMENTACIÓN (Uso por el Bibliotecario)

# Crear el objeto (Instancia)
libro1 = Libro("Cien años de soledad", "Gabriel García Márquez")

# Consultar información inicial
libro1.mostrar_info()

# Simular un préstamo
libro1.prestar()

# Consultar información después del préstamo para ver el cambio de estado
libro1.mostrar_info()

# Intentar prestarlo de nuevo para probar la lógica
libro1.prestar()
```
