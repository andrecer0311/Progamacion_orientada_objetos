```python
from abc import ABC, abstractmethod

# Clase abstracta/padre: Define la estructura base.
class PedidoOnline(ABC):
  def __init__(self, id_pedido, cliente):
    self.id_pedido = id_pedido
    self.cliente = cliente
    self.estado_pago = "Pendiente" # Estado inicial del pago
    self.estado_envio = "Procesando" # Estado inicial del envío
    self.fecha_despacho = None # Fecha en que el pedido fue despachado

  @abstractmethod
  def calcular_envio(self):
    # Método abstracto: Debe ser implementado por las clases hijas.
    pass

  @abstractmethod
  def confirmar_pago(self):
    # Método abstracto: Debe ser implementado por las clases hijas.
    pass

  # Método extra: Muestra un resumen básico del pedido.
  def mostrar_resumen_basico(self):
    print(f"Pedido ID: {self.id_pedido} | Cliente: {self.cliente} | Pago: {self.estado_pago} | Envío: {self.estado_envio}")
    if self.fecha_despacho:
      print(f"  -> Despachado el: {self.fecha_despacho}")

  # Adicional: método para despachar el pedido
  def despachar_pedido(self, fecha_despacho):
    if self.estado_pago == "Confirmado Nacional" or self.estado_pago == "Confirmado Internacional":
      self.estado_envio = "Despachado"
      self.fecha_despacho = fecha_despacho
      print(f"Pedido {self.id_pedido} despachado el {fecha_despacho}.")
    else:
      print(f"Error: El pedido {self.id_pedido} no se puede despachar sin pago confirmado.")


# Clase hija: PedidoNacional (hereda de PedidoOnline)
class PedidoNacional(PedidoOnline):
  def __init__(self, id_pedido, cliente, provincia):
    super().__init__(id_pedido, cliente) # Llama al constructor del padre
    self.provincia = provincia

  def calcular_envio(self):
    costo = 10.0 # Envío fijo para nacional
    print(f"Envío nacional a {self.provincia}: ${costo}")
    return costo

  def confirmar_pago(self):
    self.estado_pago = "Confirmado Nacional"
    print(f"Pago del pedido {self.id_pedido} confirmado como nacional.")


# Clase hija: PedidoInternacional (hereda de PedidoOnline)
class PedidoInternacional(PedidoOnline):
  def __init__(self, id_pedido, cliente, pais_destino):
    super().__init__(id_pedido, cliente) # Llama al constructor del padre
    self.pais_destino = pais_destino

  def calcular_envio(self):
    costo = 50.0 # Envío más caro para internacional
    print(f"Envío internacional a {self.pais_destino}: ${costo}")
    return costo

  def confirmar_pago(self):
    self.estado_pago = "Confirmado Internacional"
    print(f"Pago del pedido {self.id_pedido} confirmado como internacional.")


# Crear y usar un pedido nacional
print("\n--- Pedido Nacional ---")
pedido_nacional = PedidoNacional("PN002", "Marta López", "Barcelona")
pedido_nacional.mostrar_resumen_basico()
pedido_nacional.calcular_envio()
pedido_nacional.confirmar_pago()
pedido_nacional.mostrar_resumen_basico() # Ver estado actualizado

# Nuevo: Despachar el pedido nacional
pedido_nacional.despachar_pedido("2026-05-11")
pedido_nacional.mostrar_resumen_basico()

# Crear y usar un pedido internacional
print("\n--- Pedido Internacional ---")
pedido_internacional = PedidoInternacional("PI002", "Eva Green", "France")
pedido_internacional.mostrar_resumen_basico()
pedido_internacional.calcular_envio()
pedido_internacional.confirmar_pago()
pedido_internacional.mostrar_resumen_basico() # Ver estado actualizado

# Nuevo: Despachar el pedido internacional
pedido_internacional.despachar_pedido("2026-05-21")
pedido_internacional.mostrar_resumen_basico()
```
