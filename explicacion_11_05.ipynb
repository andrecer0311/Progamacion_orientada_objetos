```python
class Empleado:
  def __init__(self, nombre, salario):
    self.nombre = nombre
    self.salario = salario

  def mostrar_info(self):
    print("Empleado:" , self.nombre)
    print("Salario:" , self.salario)

class Programador(Empleado):
  def __init__(self, nombre, salario, lenguaje):
    super().__init__(nombre, salario)
    self.lenguaje = lenguaje

  def programar(self):
    print(self.nombre, "esta programando en", self.lenguaje)

class Diseñador(Empleado):
  def __init__(self, nombre, salario, herramienta):
    super().__init__(nombre, salario)
    self.herramienta = herramienta

  def diseñar(self):
    print(self.nombre, "Usa", self.herramienta)
p1 = Programador("Juan", 3000, "Python")
d1 = Diseñador("Maria", 2500, "Photoshop")
p1.mostrar_info()
p1.programar()
d1.mostrar_info()
d1.diseñar()
```
